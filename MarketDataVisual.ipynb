```python
import yfinance as yf
import json
import re
import datetime
import matplotlib.pyplot as plt
import matplotlib.colors as clr
import colorsys
```

```python
print('Start reading file')
```

```python
with open('data/Stocks.json', 'r') as stocks:
    data = stocks.read()
stocks = json.loads(data)
```

```python
print('File read, start getting data:')
```

```python
key = list(stocks.keys())[2]
title = stocks[key]
print(key)

start = "2018-01-01"
end = "2021-09-01"
```

```python
prices = yf.download(title, start = start, end = end)
#prices = yf.download(title, period = '1y')

highPrice = prices.High
lowPrice = prices.Low

print('OK')
```

```python
normalisedHighPrice = (highPrice/highPrice.iloc[0]*100)
normalisedLowPrice = (lowPrice/lowPrice.iloc[0]*100)
```

```python
plt.figure(figsize=(15, 10))
colorNames = list(clr.cnames.keys())
divider = len(colorNames) / len(title)
for symbole in title:
    c = colorNames[int((title.index(symbole) * divider) // 1)]
    plt.plot(normalisedHighPrice[symbole], label = symbole, aa = True, c = c)
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.ylabel(key)

plt.show()
```
