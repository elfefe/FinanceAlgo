```python
import yfinance as yf
import json
import re
import datetime
```

```python
print('Start reading file')
```

```python
with open('data/Stocks.json', 'r') as stocks:
    data = stocks.read()    
stocks = json.loads(data)
```

```python
print('File read, start getting data:')
```

```python
titleDict = {}
for title in stocks:
    print(title)
    stocksDict = {}
    for symbole in stocks[title]:
        #stock = yf.Ticker(symbole)
        start = "2020-03-01"
        end = "2021-03-30"
        prices = yf.download(symbole, start = start, end = end)
        if data is not None:
            try:
                dateDict = {}
                for date in prices.index:
                    marketDict = {
                        'Open': prices.loc[[date]]['Open'].item(),
                        'Close': prices.loc[[date]]['Close'].item(),
                        'High': prices.loc[[date]]['High'].item(),
                        'Low': prices.loc[[date]]['Low'].item(),
                        'Volume': prices.loc[[date]]['Volume'].item()
                    }
                    dateDict[date.strftime('%Y/%m/%d')] = marketDict
                stocksDict[symbole] = dateDict
            except:
                print("Exception, don't care much ! stock: " + symbole)
    titleDict[title] = stocksDict
```

```python
print('Write begin...')
```

```python
f = open("data/MarketData.json", "w")
f.write(json.dumps(titleDict, indent = 4))
f.close()
```
